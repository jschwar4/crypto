```python
# Initial imports
import pandas as pd
import hvplot.pandas
from path import Path
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Load the crypto_data.csv dataset.
dataset_path = "/Users/jaredschwartz/Downloads/crypto_data.csv"
crypto_df = pd.read_csv(dataset_path)
crypto_df.dtypes
```

```python
crypto_df = crypto_df.loc[crypto_df["IsTrading"]==True]
crypto_df
```

```python
crypto_df = crypto_df.loc[crypto_df["TotalCoinsMined"]>0]
crypto_df
```

```python
crypto_df = crypto_df.dropna()
crypto_df
```

```python
crypto_names = pd.DataFrame(crypto_df["CoinName"])
crypto_names.set_index(crypto_df.index)
crypto_names
```

```python
crypto_df.drop(["CoinName","Unnamed: 0","IsTrading"],axis=1,inplace=True)
crypto_df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
crypto_df["Algorithm"] = le.fit_transform(crypto_df["Algorithm"])
crypto_df["ProofType"] = le.fit_transform(crypto_df["ProofType"])
crypto_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaler = scaler.fit(crypto_df)
X = pd.DataFrame(X_scaler.transform(crypto_df))
X
```

```python
X.columns = crypto_df.columns
X.index = crypto_df.index
X
```

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
pca.explained_variance_ratio_
```

```python
pcs_df = pd.DataFrame(X_pca)
pcs_df.index = crypto_names["CoinName"]
pcs_df.columns = ["PCS 1","PCS 2","PCS 3"]
pcs_df
```

```python
inertia = []
k = list(range(1, 11))

for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)
```

```python
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```
